```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_excel('/content/Gastos y costos 20-23.xlsx')
df
```

```python
df.info()
```

```python
val_nul = df.isnull().sum()
val_nul
```

```python
df = df.drop(range(0,4))
df.columns = df.iloc[0]
df = df[1:]
df
```

```python
df['GASTO'].mode()
```

```python
df['GASTO'] = df['GASTO'].fillna('COMPRA')
```

```python
df['TC'] = df['TC'].fillna(1)
```

```python
df['IMPORTE'].median()
```

```python
df['IMPORTE'] = df['IMPORTE'].fillna(490)
```

```python
df['IVA'] = df['IVA'].fillna(df['IVA'].mean())
```

```python
df['TIPO'] = df['TIPO'].fillna('I')
```

```python
df['POLIZA'] = df['POLIZA'].replace('NO LOS METI', 'PE07-ENE20')
df['POLIZA'] = df['POLIZA'].fillna('PE07-ENE20')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df['POLIZA'].value_counts()
```

```python
df['FOLIO'].value_counts()
```

```python
df = df.drop('FOLIO', axis=1)
```

```python
df.info()
```

```python
df['RET ISR'] = df['RET ISR'].astype('float')
df['RET IVA'] = df['RET IVA'].astype('float')
df['TOTAL MX'] = df['TOTAL MX'].astype('float')
df['TOTAL SAT'] = df['TOTAL SAT'].astype('float')
```

```python
df2 = df[['TC', 'IMPORTE', 'IVA', 'RET ISR', 'RET IVA', 'TOTAL MX', 'TOTAL SAT']]
df2
```

```python
y=df
Limite_Superior= y.mean() + 3*y.std()
Limite_Inferior= y.mean() - 3*y.std()
print("Limite superior permitido", Limite_Superior)
print("Limite inferior permitido", Limite_Inferior)
```

```python
df3= df2[(y<=Limite_Superior)&(y>=Limite_Inferior)]
df3
```

```python
valores_nulos=df3.isnull().sum()
valores_nulos
```

```python
df_clean=df3.copy()
df_clean=df_clean.fillna(round(df3.mean(),1))
df_clean
```

```python
valores_nulos=df_clean.isnull().sum()
valores_nulos
```

PROCEDIMIENTO "CUANTILES" PARA SUSTITUIR OUTLIERS EN DATAFRAME

```python
y=df2

percentile25=y.quantile(0.25) #Q1
percentile75=y.quantile(0.75) #Q3
iqr= percentile75 - percentile25

Limite_Superior_iqr= percentile75 + 1.5*iqr
Limite_Inferior_iqr= percentile25 - 1.5*iqr
print("Limite superior permitido", Limite_Superior_iqr)
print("Limite inferior permitido", Limite_Inferior_iqr)
```

```python
df3_iqr= df2[(y<=Limite_Superior_iqr)&(y>=Limite_Inferior_iqr)]
df3_iqr
```

```python
valores_nulos=df3_iqr.isnull().sum()
valores_nulos
```

```python
df5_iqr=df3_iqr.copy()
df5_iqr=df5_iqr.fillna(round(df3_iqr.mean(),1))
df5_iqr
```

```python
df5_iqr.to_csv("Datos_limpios.csv")
```
